# Linear Regression in Python using SciKit Learn

```python
# load libaries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math
```

```python
# load dataset
dataset = pd.read_csv('Diwali_Sales_Data.csv')
```

```python
# Single Feature Variable - Size of the Houses
size_data = dataset['Age']
```

```python
# Target/Dependent Variable - Price of the Houses
price_data = dataset['Amount']
```

```python
# Size and Target are dataframes. We have to convert them into Array to be used as Training dataset. 
# and then use eshape function to convert array.shape(1,n) to array.shape(n,1) so each independent variable has own row.

Age = np.array(size_data).reshape(-1,1)
Amount = np.array(price_data).reshape(-1,1)
```

```python
# Remove rows where 'Amount' is NaN
cleaned_data = dataset.dropna(subset=['Amount'])

# Define features and target
Age = cleaned_data['Age'].values.reshape(-1, 1)
Amount = cleaned_data['Amount']

# Train the model
model = LinearRegression()
model.fit(Age, Amount)
```

```python
# Predict Price
price_predicted = model.predict(Age)
```

```python
# Plot the result
plt.scatter(Age,Amount, color="green")
plt.plot(Age,price_predicted, color="red")
plt.title("Linear Regression")
plt.xlabel("House Size")
plt.ylabel("House Price")
```
